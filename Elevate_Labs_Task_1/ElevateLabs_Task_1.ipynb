```python
import pandas as pd
```

```python
def Null_remover(df):
  for i in df.columns:
    if df[i].dtype=='object':
      df[i]=df[i].fillna(df[i].mode()[0])
    else:
      df[i]=df[i].fillna(df[i].mean())
  return df


from sklearn.preprocessing import LabelEncoder, OneHotEncoder
def Label(df,i):
  le=LabelEncoder()
  df[i]=le.fit_transform(df[i])
  return df[i]


def ohe(df,cols):
  df = pd.get_dummies(df,columns=cols,prefix=cols,drop_first=False)
  return df
```

```python
df1=pd.read_csv('/content/marketing_campaign.csv', delimiter='\t')
df1.info()
```

```python
print(df1.isnull().sum().sum())
df1=Null_remover(df1)
```

```python
df1.drop_duplicates(ignore_index=True,inplace=True)
```

```python
df1.columns = (df1.columns.str.strip().str.lower())
```

```python
df1=ohe(df1,['education','marital_status'])
df1['dt_customer'] = pd.to_datetime(df1['dt_customer'], errors='coerce')
```

Sales_Data

```python
df5=pd.read_csv('/content/sales_data.csv',encoding='latin1')
df5.info()
```

```python
df5.isnull().sum()
df5['ADDRESSLINE2'] = df5['ADDRESSLINE2'].fillna('Not Available')
df5['STATE'] = df5['STATE'].fillna('Unknown')
df5['POSTALCODE'] = df5['POSTALCODE'].fillna('Unknown')
df5['TERRITORY'] = df5['TERRITORY'].fillna('Unknown')
```

```python
print(df5.duplicated().sum())
df5.drop_duplicates(ignore_index=True,inplace=True)
```

```python
df5.columns = (df5.columns.str.strip().str.lower())
df5.columns
```

```python
df5['orderdate']=pd.to_datetime(df5['orderdate'],errors='coerce')
df5['phone']=pd.to_numeric(df5['phone'], errors='coerce')
df5['postalcode']=pd.to_numeric(df5['postalcode'], errors='coerce')

df5['phone']=df5['phone'].astype('Int64')
df5['postalcode']=df5['postalcode'].astype('Int64')
```
